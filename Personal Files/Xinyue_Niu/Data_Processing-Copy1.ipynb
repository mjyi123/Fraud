# Drop Unused Columns

```python
df = df[['FULLVAL', 
         'AVLAND', 
         'AVTOT', 
         'LTFRONT', 
         'LTDEPTH', 
         'BLDFRONT', 
         'BLDDEPTH', 
         'STORIES', 
         'ZIP', 
         'TAXCLASS', 
         'B',
         'BLOCK',
         'BLDGCL']]
```

```python
df.info()
```

```python
df.head()
```

Type Changing_Categotical

```python
df['ZIP'] = df['ZIP'].astype('object')
df['TAXCLASS'] = df['TAXCLASS'].astype('object').astype('category')
df['B'] = df['B'].astype('object').astype('category')
df['BLOCK'] = df['BLOCK'].astype('object').astype('category')
df['BLDGCL'] = df['BLDGCL'].astype('category')
```

```python
df.info()
```

```python
df.head()
```

## Filling NAs: FULLVAL, AVLAND, AVTOT

```python
df['AVTOT'] = df['AVTOT'].fillna(df.group(['B','BLOCK'])['AVTOT'])
```
