```python
## copy df
df_new = df
## replace 0 in 'AVTOT' with NA
df_new['AVTOT'] = df_new['AVTOT'].replace(0, np.nan)
## replace 0 in 'FULLVAL' with NA
df_new['FULLVAL'] = df_new['FULLVAL'].replace(0, np.nan)
## replace 0 in 'AVLAND' with NA
df_new['AVLAND'] = df_new['AVLAND'].replace(0, np.nan)
## Fields that need to be filled
filling_fileds = ['FULLVAL', 'AVLAND', 'AVTOT']
## Loop starts
for filling_filed in filling_fileds:
    ## Creating new df that does not have NAs in the field of filling_filed
    ## group by ['ZIP', 'BLDGCL'] to find median in each group
    right = df_new[df_new[filling_filed].notna()].\
                groupby(['ZIP', 'BLDGCL'])[filling_filed].median().reset_index()
    ## Creating new df that the the filed of filling_filed is NA
    left = df_new[df_new[filling_filed].isna()]
    ## merge right df and left df on ['ZIP', 'BLDGCL']
    result = pd.merge(left = left, 
                      right = right, 
                      how = 'left', 
                      on = ['ZIP', 'BLDGCL'], 
                      suffixes=('_left', '_right'))
    ## rename the column and drop the column that contains NA
    result = result.drop(columns = [filling_filed + '_left'])
    result = result.rename(columns = {filling_filed + '_right':filling_filed})
    ## concatenate df that do not have missing value and the result
    upper_df = df_new[df_new[filling_filed].notna()]
    df_new = pd.concat([upper_df, result], ignore_index = True, sort = True)
## Still the new df has NAs in fields of ['AVTOT', 'FULLVAL', 'AVLAND']
## So the next step is filling the NAs in these fields with the total median
median_AVTOT = np.nanmedian(df_new['AVTOT'])
## So the next step is filling the NAs in these fields with the total median
median_FULLVAL = np.nanmedian(df_new['FULLVAL'])
## So the next step is filling the NAs in these fields with the total median
median_AVLAND = np.nanmedian(df_new['AVLAND'])
## Start filling!!!!
df_new['AVTOT'] = df_new['AVTOT'].fillna(median_AVTOT)
df_new['FULLVAL'] = df_new['FULLVAL'].fillna(median_FULLVAL)
df_new['AVLAND'] = df_new['AVLAND'].fillna(median_AVLAND)
```

```python
import pandas as pd
import numpy as np
# load csv
df = pd.read_csv('NY property data.csv')
#change data type
df['ZIP'] = df['ZIP'].astype('category')
df['TAXCLASS'] = df['TAXCLASS'].astype('object').astype('category')
df['B'] = df['B'].astype('object').astype('category')
df['BLOCK'] = df['BLOCK'].astype('object').astype('category')
df['BLDGCL'] = df['BLDGCL'].astype('category')
#combine 0 and NA
df.replace(0, np.nan, inplace=True)
#fill ZIP nan
df['ZIP'] = df.groupby(['B','BLOCK']).fillna('mode')
a=df['ZIP'].isnull().sum()
#fill NAs: FULLVAL, AVLAND, AVTOT
HV=['FULLVAL', 'AVLAND', 'AVTOT']
for hvi in HV:
    df[hvi]=df.groupby(['BLDGCL','ZIP']).transform(lambda x: x.fillna(x.median()) if len(x)>=5 else x)
    df[hvi]=df.groupby('BLDGCL').transform(lambda x: x.fillna(x.median()))
b=df[HV].isnull().sum()
#Filling NAs: LTFRONT, LTDEPTH,BLDFRONT, BLDDEPTH
HP=['LTFRONT', 'LTDEPTH', 'BLDFRONT', 'BLDDEPTH']
for hpi in HP:
    df[hpi]=df.groupby('BLDGCL').transform(lambda x: x.fillna(x.median()))
    #df[hpi]=df.groupby('').transform(lambda x: x.fillna(x.median()))
c=df[HP].isnull().sum()
# Filling NAs: STORIES
df['STORIES'] = df.groupby('BLDGCL').transform(lambda x: x.fillna(x.median()))
d=df['STORIES'].isnull().sum()
# create variables
df['zip3'] = df.ZIP.astype(str).str[:3]
df['zip5'] = df.ZIP.astype(str).str[:5]
```
